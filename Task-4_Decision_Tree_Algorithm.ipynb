# Task: To Explore Decision Tree Algorithm

### For the given ‘Iris’ dataset, create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

```python
## Get the data that is necessary
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
x = iris.data
x
```

```python
x.shape
```

```python
y = iris.target
y
```

```python
y.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x,y)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(15,12))
plot_tree(clf, filled=True,rounded=True,node_ids=True,proportion=True,
          feature_names=['SepalLength(Cm)', 'SepalWidth(Cm)', 'PetalLength(Cm)', 'PetalWidth(Cm)'])
plt.show()
```
